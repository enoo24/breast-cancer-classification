```python
#drop unnecessary columns
df = df.drop(columns=['id', 'Unnamed: 32'])
```

```python
#let's see the correlation between each feature and diagnosis results
features = ['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']

for i in range(len(features)):
    if len(features)-i > 1:
        sns.scatterplot(x=features[i], y=features[i+1], data=df, hue='diagnosis')
        plt.show()
    else:
        sns.scatterplot(x=features[i], y=features[0], data=df, hue='diagnosis')
        plt.show()
```

<p>We can see in some features' correlation that there are distinct separation between 'Benign' and 'Malignant'</p>

<h3>Prepare the data for modeling</h3>

```python
#splitting our data into training and test set
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df, test_size=0.3)

print('Train Set: {0:0.2f}%'.format(100*train_df.size/df.size))
print('Test Set: {0:0.2f}%'.format(100*test_df.size/df.size))
```

```python
#scaling our features to have mean=0 and std=1
from sklearn.preprocessing import StandardScaler

X_train = train_df[features]
y_train = np.where(train_df['diagnosis'] == 'B', 0, 1)

X_test = test_df[features]
y_test = np.where(test_df['diagnosis'] == 'B', 0, 1)

scaler = StandardScaler().fit(X_train)

X_train_df = pd.DataFrame(scaler.transform(X_train), columns=features)
X_test_df = pd.DataFrame(scaler.transform(X_test), columns=features)
```

```python
X_train_df.describe()
```

```python
X_test_df.describe()
```

<h3>Modeling</h3>

```python
#using knn algorithm to models our data with various k points
from sklearn.neighbors import KNeighborsClassifier

test_scores = []
train_scores = []

for i in range(1, 25):
    knn = KNeighborsClassifier(i)
    knn.fit(X_train_df, y_train)
    
    train_scores.append(knn.score(X_train_df, y_train))
    test_scores.append(knn.score(X_test_df, y_test))
```

```python
#printing max score for training set
max_train_score = max(train_scores)
train_scores_ind = [i for i,v in enumerate(train_scores) if v == max_train_score]

print('Max train score {}% and k = {}'.format(max_train_score*100, list(map(lambda x: x+1, train_scores_ind))))
```

```python
#printing max score for testing set
max_test_score = max(test_scores)
test_scores_ind = [i for i,v in enumerate(test_scores) if v == max_test_score]

print('Max test score {}% and k = {}'.format(max_test_score*100, list(map(lambda x: x+1, test_scores_ind))))
```

<h3>Result visualization</h3>

```python
plt.figure(figsize=(12, 6))
p = sns.lineplot(range(1, 25), train_scores, marker='*', label='Train Score')
p = sns.lineplot(range(1, 25), test_scores, marker='o', label='Test Score')
```

<p>The best result is captured at k = 3, hence 3 is used for the final model</p>

```python
knn = KNeighborsClassifier(3)

knn.fit(X_train_df, y_train)
knn.score(X_test_df, y_test)
```

<h3>Model performance analysis</h3>

```python
#confusion matrix
from sklearn.metrics import confusion_matrix

y_pred = knn.predict(X_test_df)
cnf_matrix = confusion_matrix(y_test, y_pred)
p = sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap='YlGnBu', fmt='g')
plt.title('Confusion Matrix')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
```

```python
#classification report
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
#ROC - AUC
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

y_pred_proba = knn.predict_proba(X_test_df)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

plt.plot([0,1], [0,1], 'k--')
plt.plot(fpr, tpr, label='KNN')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.title('KNN (n_neighbors = 3) ROC Curve')
plt.show()

print('Area under ROC Curve:', roc_auc_score(y_test, y_pred_proba))
```

<h3>Hyper parameter optimization</h3>

```python
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(df[features]), columns=features)
y = np.where(df['diagnosis'] == 'B', 0, 1)

from sklearn.model_selection import GridSearchCV

param_grid = {'n_neighbors': np.arange(1, 50)}
knn = KNeighborsClassifier()
knn_cv = GridSearchCV(knn, param_grid, cv=5)
knn_cv.fit(X, y)

print("Best Score:" + str(knn_cv.best_score_))
print("Best Parameters: " + str(knn_cv.best_params_))
```
